```python
from graphviz import Digraph
```

```python
g = Digraph(engine='dot')

g.node('a01', color='red')
g.node('a02', color='red')
g.node('a03', color='red')
g.node('a04', color='red')
g.node('a05', color='red')

g.node('rew')
g.node('done')

g.node('s01', color='green')
g.node('s02', color='green')
g.node('s03', color='green')
g.node('s04', color='green')
g.node('s05', color='green')

g.edge('a01', 's01')
g.edge('a02', 's02')
g.edge('a03', 's03')
g.edge('a04', 's04')
g.edge('a05', 's05')

g.edge('a01', 'rew')
g.edge('a02', 'rew')
g.edge('a03', 'rew')
g.edge('a04', 'rew')
g.edge('a05', 'rew')

g.edge('s01', 'rew')
g.edge('s02', 'rew')
g.edge('s03', 'rew')
g.edge('s04', 'rew')
g.edge('s05', 'rew')

g.edge('s01', 'done')
g.edge('s02', 'done')
g.edge('s03', 'done')
g.edge('s04', 'done')
g.edge('s05', 'done')
```
